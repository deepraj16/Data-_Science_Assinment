```python
import scipy.stats as sta
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch 
```

```python
x['max']=x['max'].astype(int)
x['min']=x['min'].astype(int)
x['mean']=x['mean'].astype(int)
x.head(2)
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return x
```

```python
x=norm_func(x.iloc[:,1:])
```

```python
z=linkage(x,metric="euclidean")
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
h_com=AgglomerativeClustering(n_clusters=3,linkage='complete',affinity="euclidean").fit(x)
```

```python
plt.figure(figsize=(9,4));
# optional part
plt.title("Hierical clusring")
plt.xlabel("Index");
plt.ylabel("distance")
#requried part:
sch.dendrogram(z,leaf_rotation=0,leaf_font_size=10)
plt.show()
```
