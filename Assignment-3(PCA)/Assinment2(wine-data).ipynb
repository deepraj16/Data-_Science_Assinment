```python
from sklearn.datasets import load_wine 
import pandas as pd
import numpy as np
```

```python
X,y=load_wine(return_X_y=True)
```

```python
y
```

```python
pd.DataFrame(X)
```

```python
lm =pd.read_csv('wine.csv.xls')
lm.head(3)
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestClassifier 
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
```

```python
clf1 =DecisionTreeClassifier()
clf2=LogisticRegression() 
clf3=RandomForestClassifier()
x_train,x_test,y_train,y_test =train_test_split(X,y,random_state=42,test_size=0.2)
clf1.fit(x_train,y_train)
clf2.fit(x_train,y_train)
clf3.fit(x_train,y_train)
y_pre1 = clf1.predict(x_test)
y_pre2 = clf2.predict(x_test)
y_pre3 = clf3.predict(x_test)
print("Decision tree score : ",accuracy_score(y_test,y_pre1)) 
print("Logistic regression score : ",accuracy_score(y_test,y_pre2)) 
print("Random forest score : ",accuracy_score(y_test,y_pre3)) 
```
